## 1a )  Generating Spatial Gratings having sinusoidal amplitudes. 

Itâ€™s called a sinusoidal grating as the grayscale amplitude in the grating vary according to sine function.
Three parameters that describe the grating :
1. Amplitude (Contrast of the grating)
2. Wavelength / Frequency (Spatial Frequency)
3. Phase (For comparison)

Here, we only vary the spatial frequency as required, keeping the other two parameters constant. 

We will use numpy package to create our functions and matplotlib for visualiasations of gratings. 

Finally, to define 2D Grating, we use np.meshgrid() for 2D representation. Here X and Y  represent the 2D axes.

The x array contains the horizontal scale, defined in micrometer (10^-6 m) range. 

#####  1. Spatial Freq = 10 lines/mm 
Wavelength ($\lambda$) = $1/\nu$ = 100 $\mu$m

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt

#creating an array with 10001 elements to represent the x-axis scale
x = np.arange(-500, 501, 1)

wavelength = 100
y1 = np.sin(2 * np.pi * x / wavelength)
#creating our y = sin(x) function

plt.plot(x, y1)
plt.xlabel("in micometer scale")
plt.ylabel("Normalised Amplitude")
plt.show()
```

```python
X, Y = np.meshgrid(x, x)

wavelength = 100
grating_1 = np.sin(2 * np.pi * X / wavelength)

#Setting the colour of sinusoidal variation to gray, to represent our grating
plt.set_cmap("gray")
plt.imshow(grating_1)
plt.title("10 lines/mm")
plt.xlabel("in micometer scale")
plt.show()
```

## 1 b)  Adding the 4 distributions. 
Directly adding the amplitudes of the 4 sine functions with the 4 spatial frequecies, since they are already in phase. 

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt

#creating an array with 10001 elements to represent the x-axis scale
x = np.arange(-500, 501, 1)

#Adding the 4 spatial frequencies
y= y1+y2+y3+y4


plt.plot(x, y)
plt.title("Resultant Amplitude Distribution")
plt.xlabel("in micometer scale")
plt.ylabel("Amplitude (can be normalised)")
plt.show()


```

```python
X, Y = np.meshgrid(x, x)

grating = grating_1 + grating_2 +grating_3 + grating_4

#Setting the colour of sinusoidal variation to gray, to represent our grating
plt.set_cmap("gray")
plt.imshow(grating)
plt.title("Resultant Grating")
plt.xlabel("in micometer scale")
plt.show()
```

## 1.c) Inverse Fourier Transform to obtain the original 4 gratings.  

```python
ft = np.fft.fft(y)
abs(ft)
#We use the fft 
```

```python
## The function below calculates the 2D Fourier transform in Python
ft = np.fft.ifftshift(grating)
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)
## The above two call lines, swaps the image pixel array. This ensures the distance of points in FFT Plot from this centre contains the frequency of the sinusoidal gratings.



plt.figure()
plt.subplot(121)
plt.imshow(grating)
plt.subplot(122)

##  We plot the absolute value of the Fourier transform ft, using the abs() built-in function
plt.imshow(abs(ft))

## Setting the limit for the required range of scale (10 lines/mm to 70 lines/mm).  
plt.xlim([420, 580])
plt.ylim([580, 420])  # Note, order is reversed for y
plt.show()
```

On the left, is the superposed grating of the 4 spatial frequencies (10 lines/mm, 30 lines/mm, 50 lines/mm, 70 lines/mm)                  
Meanwhile on the right side is the FT of the superposed sinusoidal funtion, giving us back the original spatial frequencies. 
**The distance of the dots from the centre represents the frequency of the sinusoidal grating**. The further the dots are from the centre, the higher the frequency they represent.

**The centre of the image is at 500,500 pixel position. The dots are at distances 10, 30, 50, 70 pixels, from the centre. Thereby giving us back the original the spatial frequencies of 10, 30 , 50 and 70 lines/mm, which can be plotted to get back as in (a).**


Note : Also, The value of the pixels making up the dots in the Fourier transform represents the amplitude of the grating. 
Therefore, we can generate the original resultant grating by taking inverse Fourier Transform of this.
